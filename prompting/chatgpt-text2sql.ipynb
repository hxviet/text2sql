```python
import json
dev_data = json.load(open('/kaggle/input/spider/spider/dev.json'))



table_data = json.load(open('/kaggle/input/spider/spider/tables.json'))

data_into_prompt = {}
for entry in table_data:
    db_id = entry['db_id']
    data_into_prompt[db_id] = []
    for i in range(len(entry['table_names_original'])):
        nametup = (entry['table_names_original'][i],[])
        for colnm in entry['column_names_original']:
            if colnm[0] == i:
                nametup[1].append(colnm[1])
        data_into_prompt[db_id].append(nametup)
```

```python
import time
import random as rd
```

```python
for data_sample in dev_data[:2]:
    db_id = data_sample['db_id']
    prompt_info = ''
    for tbl in data_into_prompt[db_id]:
        prompt_info += f'{tbl[0]} with columns {", ".join(tbl[1])}\n'
    prompt_entry = f'Provide a SQL query corresponding to this question in the bracket: ({data_sample["question"]})\n\
    given that the database has {len(data_into_prompt[db_id])} tables:\n{prompt_info}'


    response = g4f.ChatCompletion.create(model='gpt-3.5-turbo', messages=[
                                     {"role": "user", "content": prompt_entry}], stream=False)
    res = response.replace("\n"," ")
    idz = res.find("SELECT")
    ide = res[idz:].find(";")
    res = res[idz:idz+ide]

    print(f"Prompt: {prompt_entry}")
    print("######################")
    print(f"Result: {res}")

    print("\n\n\n")

    time.sleep(60) #this to ensure that we don't exceed the quota
```
